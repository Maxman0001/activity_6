```python
import os
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = (
    SparkSession.builder
    .appName("HealthSurveyProcessing")
    .getOrCreate()
)
```

```python
(survey :=
 spark.read
      .option("header", True)
      .option("inferSchema", True)
      .csv("/content/health_survey_v2.csv")
).limit(5).toPandas()
```

```python
(survey :=
 spark.read
      .option("header", True)
      .option("inferSchema", True)
      .csv("/content/health_survey_v2.csv")
).limit(5).toPandas()
```

```python
(reverse :=
 spark.read
      .option("header", True)
      .option("inferSchema", True)
      .csv("/content/ReverseCodingItems_v2.csv")
).limit(5).toPandas()
```

Create coding maps

```python
reg = {
    "Strongly Agree": 5,
    "Somewhat Agree": 4,
    "Neither Agree nor Disagree": 3,
    "Somewhat Disagree": 2,
    "Strongly Disagree": 1
}

rev = {
    "Strongly Agree": 1,
    "Somewhat Agree": 2,
    "Neither Agree nor Disagree": 3,
    "Somewhat Disagree": 4,
    "Strongly Disagree": 5
}

reg_map = F.create_map([F.lit(x) for kv in reg.items() for x in kv])
rev_map = F.create_map([F.lit(x) for kv in rev.items() for x in kv])
```

Stack (UNPIVOT)

```python
q_cols = [c for c in survey.columns if c != "ID"]

stack_expr = " , ".join(
    [f"'{c}', `{c}`" for c in q_cols]
)

(long_df :=
 survey.select(
     "ID",
     F.expr(f"stack({len(q_cols)}, {stack_expr}) AS (Question, Response)")
 )
).limit(10).toPandas()
```

Join Reverse Coding Table (FIX AMBIGUITY)

```python
(joined :=
 long_df
 .join(reverse, long_df.Question == reverse["Column Name"], "left")
 .select(
     long_df.ID,
     long_df.Question,
     long_df.Response,
     reverse["Needs Reverse Coding?"]
 )
).limit(10).toPandas()
```

Add Temp Regular + Temp Reverse Values

```python
(coded1 :=
 joined
 .withColumn("TempReg", reg_map[F.col("Response")].cast("int"))
 .withColumn("TempRev", rev_map[F.col("Response")].cast("int"))
).limit(10).toPandas()
```

Final Recoded Value (CASE-WHEN)

```python
(coded2 :=
 coded1.withColumn(
     "RecodedValue",
     F.when(F.col("Needs Reverse Coding?") == "Yes",
            F.col("TempRev")
     ).otherwise(F.col("TempReg"))
 )
).limit(10).toPandas()
```

Extract Question Type (F1, F2, …)

```python
(coded2 :=
 coded1.withColumn(
     "RecodedValue",
     F.when(F.col("Needs Reverse Coding?") == "Yes",
            F.col("TempRev")
     ).otherwise(F.col("TempReg"))
 )
).limit(10).toPandas()
```

```python
(coded3 :=
 coded2.withColumn(
     "QType",
     F.split("Question", "\\.").getItem(0)
 )
).limit(10).toPandas()
```

Pivot (Aggregate & Split)

```python
(final_pivot :=
 coded3
 .groupBy("ID")
 .pivot("QType")
 .agg(F.sum("RecodedValue"))
 .orderBy("ID")
).limit(10).toPandas()
```

```python
(final_pivot :=
 coded3
 .groupBy("ID")
 .pivot("QType")
 .agg(F.sum("RecodedValue"))
 .orderBy("ID")
).limit(10).toPandas()
```

```python
final_output_path = "/content/health_survey_summary"

final_pivot.coalesce(1).write.option("header", True).mode("overwrite").csv(final_output_path)

final_output_path
```
